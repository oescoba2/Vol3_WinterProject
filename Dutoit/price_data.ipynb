```python
import pandas as pd

data = pd.read_csv('data.csv')
print(data.head())

food_prices = data[data['Product'] == 'food_price_index']
print(food_prices.head())
```

```python
prices = food_prices[["Date", "Close"]]
print(prices.head())
```

```python
from matplotlib import pyplot as plt
import numpy as np
plt.plot(prices["Date"], prices["Close"])
plt.show()
```

```python
from scipy.optimize import curve_fit
import numpy as np

def exponential(x, a, b, c):
    return  a*b**x + c

x = np.arange(len(prices))
y = prices["Close"].values

popt, _ = curve_fit(exponential, x, y, p0=[1, 1, 1])

plt.plot(x, prices["Close"])
plt.plot(x, exponential(x, *popt))
plt.grid()
plt.show()

residuals = y / exponential(x, *popt) 
plt.plot(x, residuals)
plt.grid()
plt.show()
```
